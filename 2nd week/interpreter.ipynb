```python
operator={
    "+": lambda x, y : (x+y),
    "-": lambda x, y : (x-y),
    "*": lambda x, y : (x*y),
    "/": lambda x, y : (x/y),
}

firstinput=str(input("Expression: ")).split(" ")
x, y, z=[i for i in firstinput if i != " "]

print(operator[y](float(x), float(z)))
```

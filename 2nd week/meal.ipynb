```python
def main():
    time=convert(str(input("What time is it? ")))
    
    if time>= 7 and time<=8:
        print("breakfast time")
    elif time>= 12 and time<=13:
        print("lunch time")
    elif time>= 18 and time<=19:
        print("dinner time")

def convert(time):
    hour, min=time.split(":")
    return float(hour) + 100/60*float(min)*0.01


if __name__ == "__main__":
    main()
```
